## algoritma k-means

```python
def euc_distance(x,y):
  return (sum((x-y)**2))**0.5
```

```python
#cluster_head
cluster_indices= np.random.choice(len(df_numpy), size=3, replace=False)
cluster_heads=[]
for i in cluster_indices:
    cluster_heads.append(df_numpy[i])
cluster_heads=np.array(cluster_heads)

cluster_heads
```

```python
def nearest_centroid(ic, data):
    assigned_centroid = []
    for i in data:
        distance=[]
        for j in ic:
            distance.append(euc_distance(i, j))
        assigned_centroid.append(np.argmin(distance))
    return(assigned_centroid)
```

```python
nilai_cluster = nearest_centroid(cluster_heads, df_numpy)
cluster = pd.DataFrame(nilai_cluster)
```

```python
def hitung_centroid(cluster, data):
  centroid_baru={}
  value_cen=[]
  for i in range(len(nilai_cluster)):
      if nilai_cluster[i] not in centroid_baru.keys():
          centroid_baru[nilai_cluster[i]]=[data[i]]
      else:
          centroid_baru[nilai_cluster[i]].append(data[i])
  for key in centroid_baru.keys():
      value_cen.append(np.mean(centroid_baru[key],axis=0))
  return value_cen

```

```python
cluster_baru = []

for i in range (20):
  nilai_cluster = nearest_centroid(cluster_heads, df_numpy)
  nilai_centroid = hitung_centroid(nilai_cluster, df_numpy)
  cluster_baru = nearest_centroid(nilai_centroid, df_numpy)
  if nilai_cluster==cluster_baru:
      break
  else:
      cluster_heads=nilai_centroid
      #print(cluster_baru)
      continue
```
